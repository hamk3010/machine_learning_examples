```python
import re
import numpy as np
import matplotlib.pyplot as plt

X=[]
Y=[]

non_decimal = re.compile(r'[^\d]+')

for line in open('moore.csv'):
    r = line.split('\t') #split on tabs
    x= int(non_decimal.sub('', r[2].split('[')[0]))
    y= int(non_decimal.sub('', r[1].split('[')[0]))
    
    X.append(x)
    Y.append(y)
    

X= np.array(X)
Y= np.array(Y)

plt.scatter(X,Y)
plt.show()

Y = np.log(Y)
plt.scatter(X,Y)
plt.show()


    
```

```python
denominator= (X.dot(X) - X.mean()*X.sum() )
a = (X.dot(Y) - Y.mean()*X.sum())/ denominator
b = (Y.mean()* X.dot(X) - X.mean()*X.dot(Y) )/ denominator

Yhat = a*X + b
plt.scatter(X,Y)
plt.plot(X,Yhat)
plt.show()

residual= Y- Yhat
total = Y - Y.mean()

SS_residual = residual.dot(residual) ##Sum of squared Residual
SS_total = total.dot(total) ##Sum of total reisdual (mean line)

r_sqaured = 1 - (SS_residual/SS_total)
```
