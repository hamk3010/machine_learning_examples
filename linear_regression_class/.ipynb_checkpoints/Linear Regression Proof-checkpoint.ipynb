```python
import numpy as np
import matplotlib.pyplot as plt

#load data
X=[]
Y= []

for line in open('data_1d.csv'):
    x,y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
    
X=np.array(X)
Y= np.array(Y)

plt.scatter(X,Y)
plt.show()
```

```python
#apply the equation hat we learned to calculate a and b
denom= X.dot(X) - X.mean()*X.sum()
# denom= X.dot(X)/len(X) - X.mean()*X.mean()

a= (X.dot(Y) - Y.mean()*X.sum())/denom

b= (Y.mean()*X.dot(X) - X.mean()*X.dot(Y) )/ denom
```

```python
Yhat = a*X + b

plt.scatter(X,Y)
plt.plot(X,Yhat)
plt.show()
```

```python
#Solving for r_squared
SS_residual = Y - Yhat #this is the error between the real Y point and our line of best fit
SS_tot= Y - Y.mean()
# for element in Y:
#     SS_tot.append(float(element-Y.mean()) )

r_squared= 1 - (SS_residual.dot(SS_residual) / SS_tot.dot(SS_tot) )
print "r_sqaured= "+ str(r_squared)
```
