```python
import numpy as np
import matplotlib.pyplot as plt

#load the data 
X=[]
Y=[]

for line in open('data_poly.csv'):
    x,y = line.split(',')
    x = float(x)
    X.append([1,x, x*x] )
    Y.append(float(y) )
    
#convert to numpy arrays
X= np.array(X)
Y= np.array(Y)

plt.scatter(X[:,1], Y)
plt.show()
```

```python
plt.scatter(X[:,2], Y)
plt.show()
```

```python
w = np.linalg.solve( np.dot(X.T,X) , np.dot(X.T,Y) )
Yhat = np.dot(X,w)

plt.scatter(X[:,1], Y)
plt.plot(sorted(X[:,1]) , sorted(Yhat) )
plt.show()
```
